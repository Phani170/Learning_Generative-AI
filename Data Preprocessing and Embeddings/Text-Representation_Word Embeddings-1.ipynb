# One Hot Encoding

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.feature_extraction.text import CountVectorizer
df = pd.DataFrame({"text":["people watch dsa",
                         "dsa watch dsa",
                         "people write comment",
                          "dsa write comment"],"output":[1,1,0,0]})
```

```python
# Initialize CountVectorizer with binary=True (one-hot style)
vectorizer = CountVectorizer(binary=True)
```

```python
# Fit and transform the text column
X = vectorizer.fit_transform(df['text'])
```

```python
# Convert to DataFrame
one_hot_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

# Combine with the original 'output' column
final_df = pd.concat([df[['text', 'output']], one_hot_df], axis=1)
```

#  N-grams

```python
df = pd.DataFrame({"text":["people watch dsa",
                         "dsa watch dsa",
                         "people write comment",
                          "dsa write comment"],"output":[1,1,0,0]})

df
```

```python
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(2,2))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
#Ti gram
# BI grams
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(ngram_range=(3,3))
```

```python
bow = cv.fit_transform(df['text'])
```

```python
print(cv.vocabulary_)
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```
